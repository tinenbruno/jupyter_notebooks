# Pandas usage example with Iris Dataset

The main focus of this document is to present ways in which some functions of pandas can be used to easily work with datasets, in terms of filtering, grouping and graphing.

## Introduction

First of all we need to import the relevant libraries that will be explored. Of course, pandas will be imported. The [Seaborn](http://seaborn.pydata.org/index.html) library will be used to do some plotting of the data.

```python
import pandas as pd
import seaborn as sns
```

The next line is needed for inline jupyter documents ploting of graphs.

```python
%matplotlib inline
```

First of all we need to load our dataset into a Pandas Dataframe. We will load it directly from the online UCI archive. Loading with `read_table` in this case would be equivalent to using `read_csv`, due to the separator used.

```python
iris = pd.read_table("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data", sep=',')
iris.head()
```

Note that in this case the dataset has no header. To better use it, let's add one based on the dataset description, using both the `header` named parameter to inform that the dataset has no inherent header and the `names` attribute to inform the names of the columns.

```python
headers = ['PetalLengthCm', 'PetalWidthCm', 'SepalLengthCm', 'SepalWidthCm', 'Species']
iris = pd.read_table("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",
                     sep=',',
                     header=None,
                     names=headers)
iris.head()
```

We can then easily see the distribution of the instances over the attributes using a pairwise plot bundled in Seaborn.

```python
sns.set()

sns.pairplot(iris, hue="Species", diag_kind='kde')
```
